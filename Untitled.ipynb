```python
def Q_learning(env,num_episodes,Q,alpha,epsilon,gamma,lamb):
    l = []    
    for _ in range(num_episodes):
        total_score = 0
        eligibility=np.zeros(Q.shape)#define eligibility
        #epsilon = epsilon * 0.9
        observation = env.reset()# reset the environment to start
        last_state = observation#set last state
        r_list = []
        while(True):
             
            Q_old = deepcopy(Q)
            x = np.random.binomial(n=1,size=1,p=1-epsilon)
            action = -1
            if x[0] == 1:#exploitation
                action = np.argmax(Q[observation[0],observation[1],:])
            else:#exploration
                action = np.random.randint(low=0,high=4)

            old_observation = deepcopy(observation)#state s
            ##update the eligibility for the last state action
            eligibility[old_observation[0],old_observation[1],action] += 1
            observation, reward, done, info = env.step(action)#take action
            
            #if reward == 0:
            #    reward = reward-10
            total_score += max(reward,0)
            r_list.append(reward)
            """
            Q-learning update step
            """
            #Q[old_observation[0], action, old_observation[1]] = \
            #    Q[old_observation[0], action, old_observation[1]] + alpha * (
            #        reward + gamma*np.max(Q[observation[0], :, observation[1]]) - Q[
            #            old_observation[0], action, old_observation[1]]) 
            delta = reward + gamma*np.max(Q[observation[0], observation[1], :]) - Q[
                        old_observation[0], old_observation[1],action]
            Q += alpha * delta * eligibility  
            eligibility *= lamb

            if done:
                #print("Target set = " + str(total_score))
                l.append(total_score)
                break
    #plt.plot(np.array(range(len(l))),np.array(l))
    #plt.show()
    return l
```

```python
max_darts = 6
max_lives = 2
num_actions = 4
```

```python
total_states = (max_darts + 1) * (max_lives + 1) #since two features
T = np.zeros((total_states,num_actions,total_states)) #state transition matrix(P(s'|s,a))
#now T is a 3d matrix we will reshape it later for calculation purposes
pi = np.ones((total_states,num_actions))/num_actions #equiprobable policy
V = np.zeros((total_states,1))#V values
R = np.zeros((total_states,num_actions))#expected reward r(s,a)
```

```python
success_prob = {0:0.95,1:0.88,2:0.80,3:0.6}
action_max_reward = {0:1,1:2,2:4,3:6}
action_expected_reward =  {action:action_max_reward[action]*success_prob[action] 
                           for action in env.action_space}
```

```python
#set T
def initialize_T():
    for action in env.action_space:
        print(action)
        for state in range(total_states):
            time_elapsed = state//(max_lives + 1)
            lives_lost = state % (max_lives+1)
            if lives_lost == max_lives or time_elapsed == max_darts:#if state i terminal
                continue
            if lives_lost > time_elapsed:
                continue
            next_time_elapsed = time_elapsed + 1
            next_state_success = next_time_elapsed * (max_lives + 1) + lives_lost
            next_state_failure = next_time_elapsed * (max_lives + 1) + (lives_lost + 1) 
            T[state,action,next_state_success] = success_prob[action]
            T[state,action,next_state_failure] = 1 - success_prob[action]
            R[state,action] = action_expected_reward[action]
            #input(str(time_elapsed) + " "
            #      + str(lives_lost) + " -->" 
            #      + str(next_state_success//(max_lives + 1)) + " " + 
            #     str(next_state_success%(max_lives + 1)) + " "  + str(T[state,action,next_state_success]))

            #input(str(time_elapsed) + " "
            #      + str(lives_lost) + " -->" 
            #      + str(next_state_failure//(max_lives + 1)) + " " + 
            #     str(next_state_failure%(max_lives + 1)) + " "  + str(T[state,action,next_state_failure]))
```

```python
#calculate T_pi
def calc_T_pi(T,pi):
    
    T = T.reshape((total_states*num_actions,total_states))#change shape to (s*a,s)
    pi = pi.reshape((total_states*num_actions,1))#reshape to (s*a,1)
    pi.shape,T.shape
    T_pi = pi*T
    T_pi.shape
    T_pi = T_pi.reshape((total_states,num_actions,total_states))
    T_pi = np.sum(T_pi,axis=1)
    T = T.reshape((total_states,num_actions,total_states))#change shape to (s,a,s)
    pi = pi.reshape((total_states,num_actions))#reshape to (s,a)
    T_pi.shape
    return T_pi
```

```python
#R_pi
def calc_R_pi(pi):
    pi = pi.reshape((total_states,num_actions))#reshape to (s,a)
    R_pi = np.sum(pi * R,axis=1)
    R_pi.shape
    return R_pi
```

```python
#update policy pi
#for every valid state 
#for all actions check (prob_success(r+v(next_success_state) + 1-prob-success(0+v(next_state_failure)))
T_pi = calc_T_pi(T,pi)
R_pi = calc_R_pi(pi)
I = np.identity(total_states)
old_V = V 
Q = np.zeros((max_darts+1,max_lives + 1,num_actions))
V = np.dot(np.linalg.inv(I - gamma * T_pi),R_pi)
for state in range(total_states):
    time_elapsed = state//(max_lives + 1)
    lives_lost = state % (max_lives+1)
    if lives_lost == max_lives or time_elapsed == max_darts:#if state i terminal
        continue
    if lives_lost > time_elapsed:
        continue
    next_time_elapsed = time_elapsed + 1
    next_state_success = next_time_elapsed * (max_lives + 1) + lives_lost
    next_state_failure = next_time_elapsed * (max_lives + 1) + (lives_lost + 1) 
    max_G = 0
    max_action = 0
    for action in env.action_space:
        G = success_prob[action]*(action_max_reward[action] + V[next_state_success]) + (1-success_prob[action])*(0 + V[next_state_failure])
        Q[time_elapsed,lives_lost,action] = G
        if max_G < G:
            max_G = G
            max_action = action
    pi[state,:] = 0
    pi[state,max_action] = 1
print(np.max(np.abs(old_V - V)))

```
